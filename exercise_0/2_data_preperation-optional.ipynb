```python
from exercise_code.data_utils import load_CIFAR10
# Load the raw CIFAR-10 data
cifar10_dir = '../datasets/'
X, y = load_CIFAR10(cifar10_dir)
print(X.shape, y.shape)
```

```python
# Your visualization code here...
def visualize(images):
    width, height = len(images[0]), len(images)
    fig = plt.figure(figsize=(width, height))
    for i in range(height):
        for j in range(width):
            ax = fig.add_subplot(height, width, i * width + j + 1)
            ax.axis('off')
            plt.imshow(images[i][j].astype(np.uint8))
    plt.show()

classes = [clazz[0][0:10] for clazz in [np.where(y==label) for label in range(10)]]
images = [X[label] for label in classes] 
visualize(images)
```
